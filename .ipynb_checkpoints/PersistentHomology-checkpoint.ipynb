```python
x=input("The pdb files: 1bh8, 1bh9, 1uao, 4f5s_1, 6mzd, 7djin, lysozyme_md. Enter the database: ")
path="./data/"+x+".pdb"
ppdb_df = PandasPdb().read_pdb(path)
#get the pandas data frame
ppdb = PandasPdb()
ppdb.read_pdb(path)
ppdb.df['ATOM'].head()
```

```python
#extact the x,y,z coordinates of the backbone atom from the dataframe as numpy.ndarray
v=ppdb.df['ATOM'][(ppdb.df['ATOM']['atom_name'] == 'C') | 
                            (ppdb.df['ATOM']['atom_name'] == 'O') | 
                            (ppdb.df['ATOM']['atom_name'] == 'N') | 
                            (ppdb.df['ATOM']['atom_name'] == 'CA')]
coord=v[['x_coord','y_coord','z_coord']]
coord2=v[['x_coord','y_coord','z_coord','b_factor']]
backbone=coord.to_numpy()
D=coord2.to_numpy()

np.savetxt("prion_yeast.txt", backbone)
```

```python
from gtda.plotting import plot_point_cloud
plot_point_cloud(backbone)
```

```python
from gtda.homology import VietorisRipsPersistence

# Track connected components, loops, and voids
homology_dimensions = [0, 1, 2]

# Collapse edges to speed up H2 persistence calculation!
persistence = VietorisRipsPersistence(
    metric="euclidean",
    homology_dimensions=homology_dimensions,
    n_jobs=6,
    collapse_edges=True,
)

diagrams_basic = persistence.fit_transform(backbone[None, :, :])
```

```python
from gtda.plotting import plot_diagram
plot_diagram(diagrams_basic[0])
```

# K-NN Distance

```python
rips = Rips(maxdim=2,thresh=20)
matrix = compute_kNN_distance(backbone, 10)
dgms = rips.fit_transform(matrix, distance_matrix=True)
rips.plot(dgms, lifetime=True)
#plt.savefig('img/k-NN_distance_(backbone).png', dpi=300, bbox_inches='tight')
plt.title('k-NN distance (backbone)');
```

```python
rips = Rips(maxdim=2,thresh=20)
matrix = compute_kNN_distance(D, 10)
dgms = rips.fit_transform(matrix, distance_matrix=True)
rips.plot(dgms, lifetime=True)
#plt.savefig('img/k-NN_distance_(with b factor).png', dpi=300, bbox_inches='tight')
plt.title('k-NN distance (with b_factor)');
```

# DTM

```python
m = 0.01                   
DTM_values = DTM(backbone,backbone,m)             

# plot of the DTM
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
ax.scatter(backbone[:,0], backbone[:,1], backbone[:,2], c=DTM_values, s = 10)
#plt.savefig('img/Values-of-the-DTM-on-X-with-parameter_m=0.01.png', dpi=300, bbox_inches='tight')
ax.set_title('Values of the DTM on X with parameter m='+str(m))
```

```python
p = 2
dimension_max = 3

# compute the persistence diagram
simplex_tree_DTM_alpha = AlphaDTMFiltration(backbone, m, p, dimension_max)
diagram_DTM_alpha = simplex_tree_DTM_alpha.persistence()    

# plot the persistence diagram
rips = Rips()
rips.plot(gudhi_to_ripser(diagram_DTM_alpha), lifetime = True)
#plt.savefig('img/DTM-filtration_(backbone).png', dpi=300, bbox_inches='tight')
plt.title('DTM-filtration (backbone)');
```

```python
p = 2
dimension_max = 3

# compute the persistence diagram
simplex_tree_DTM_alpha = AlphaDTMFiltration(D, m, p, dimension_max)
diagram_DTM_alpha = simplex_tree_DTM_alpha.persistence()    

# plot the persistence diagram
rips = Rips()
rips.plot(gudhi_to_ripser(diagram_DTM_alpha), lifetime = True)
#plt.savefig('img/DTM-filtration_(backbone).png', dpi=300, bbox_inches='tight')
plt.title('DTM-filtration (backbone with b factor)');
```
